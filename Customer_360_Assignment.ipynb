```python
# Convert date columns
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])

# Handle missing values
ratings['review_score'].fillna(ratings['review_score'].mean(), inplace=True)
payments['payment_value'].fillna(0, inplace=True)
```

```python
period1 = orders[(orders['order_purchase_timestamp'] >= '2016-11-01') &
                 (orders['order_purchase_timestamp'] <= '2017-10-31')]

period2 = orders[(orders['order_purchase_timestamp'] >= '2017-11-01') &
                 (orders['order_purchase_timestamp'] <= '2018-10-31')]
```

```python
# Merge everything step by step
df = orders.merge(items, on='order_id', how='left') \
           .merge(payments, on='order_id', how='left') \
           .merge(ratings[['order_id', 'review_score']], on='order_id', how='left')
```

```python
df['review_score'] = pd.to_numeric(df['review_score'], errors='coerce')
```

```python
df['review_score'].fillna(df['review_score'].mean(), inplace=True)
```

```python
cust_360 = df.groupby('customer_id').agg({
    'order_id': 'nunique',
    'payment_value': 'sum',
    'review_score': 'mean',
    'order_purchase_timestamp': 'max'
}).reset_index()

cust_360.columns = ['customer_id', 'total_orders', 'total_spend', 'avg_review', 'last_purchase']
```

```python
df.dtypes
```

```python
# Recency
current_date = df['order_purchase_timestamp'].max()

cust_360['recency'] = (current_date - cust_360['last_purchase']).dt.days

# Frequency already = total_orders
cust_360['frequency'] = cust_360['total_orders']

# Monetary
cust_360['monetary'] = cust_360['total_spend']
```

```python
# If no purchase in last 180 days → churn
cust_360['churn'] = cust_360['recency'].apply(lambda x: 1 if x > 180 else 0)
```

```python
threshold = cust_360['monetary'].quantile(0.75)
cust_360['high_spender'] = cust_360['monetary'].apply(lambda x: 1 if x > threshold else 0)
```

```python
cust_360
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Features
cust_360 = cust_360.copy()

cust_360['avg_order_value'] = cust_360['monetary'] / cust_360['frequency']
cust_360['avg_order_value'] = cust_360['avg_order_value'].replace([np.inf, -np.inf], 0)
cust_360['avg_order_value'] = cust_360['avg_order_value'].fillna(0)

X = cust_360[['recency', 'frequency', 'avg_review', 'avg_order_value']]

# Target (log transformed)
y = np.log1p(cust_360['monetary'])

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
model = LinearRegression()
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)

# Evaluation (log scale)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE:", rmse)
```

```python
X = cust_360[['frequency', 'monetary', 'avg_review']]
y = cust_360['churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model_log = LogisticRegression()
model_log.fit(X_train, y_train)

y_pred = model_log.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Recency score (lower recency = better customer)
cust_360['R_score'] = pd.qcut(cust_360['recency'], 4, labels=[4,3,2,1])

# Frequency score (higher = better)
cust_360['F_score'] = pd.qcut(cust_360['frequency'].rank(method='first'), 4, labels=[1,2,3,4])

# Monetary score (higher = better)
cust_360['M_score'] = pd.qcut(cust_360['monetary'], 4, labels=[1,2,3,4])
```

```python
cust_360['RFM_score'] = cust_360['R_score'].astype(str) + \
                        cust_360['F_score'].astype(str) + \
                        cust_360['M_score'].astype(str)
```

```python
def segment_customer(row):
    if row['R_score'] == 4 and row['F_score'] >= 3 and row['M_score'] >= 3:
        return 'High Value Customers'
    
    elif row['R_score'] >= 3 and row['F_score'] >= 2:
        return 'Loyal Customers'
    
    elif row['R_score'] <= 2 and row['F_score'] >= 2:
        return 'At Risk Customers'
    
    else:
        return 'Low Value Customers'

cust_360['segment'] = cust_360.apply(segment_customer, axis=1)
```

```python
segment_summary = cust_360.groupby('segment')[['recency', 'frequency', 'monetary']].mean()
print(segment_summary)
```
